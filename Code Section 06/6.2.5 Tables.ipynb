```python
import requests
import urllib.request
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
from urllib.request import urlopen
import re
```

```python
url = 'https://en.wikipedia.org/wiki/Epidemiology_of_depression'
html = urlopen(url) 
soup = BeautifulSoup(html, 'html.parser')
```

```python
soup
```

```python
tables = soup.find_all('table')
```

```python
def process_num(num):
    return float(re.sub(r'[^\w\s.]','',num))
```

```python
num1 = float(re.sub(r'[^\w\s.]','','1,156.30'))
num1
```

```python
ranks = []
rates = []
countries = []

for table in tables:
    rows = table.find_all('tr')
    
    for row in rows:
        cells = row.find_all('td')
        
        
        if len(cells) > 1:
            rank = cells[0]
            ranks.append(int(rank.text))
            
            country = cells[1]
            countries.append(country.text.strip())
            
            rate = cells[2]
            rates.append(process_num(rate.text.strip()))
            
```

```python
data = pd.DataFrame(ranks, index= countries ,columns = ['Rank'])
data['DALY rate'] = rates
data.head(10)
```

```python
data.to_csv('6.2.5 ThisTable.csv')
```
