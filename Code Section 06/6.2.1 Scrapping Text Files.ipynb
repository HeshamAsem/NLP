```python
def DownloadText(Link,FileName) : 
    content = requests.get(Link)
    print(content.text[:100])
    f = open(FileName, "w", encoding="utf-8")
    f.write(content.text)
    f.close()   
```

```python
DownloadText("http://www.gutenberg.org/cache/epub/11/pg11.txt",'6.2.1 alice.txt')
```

```python
DownloadText("https://en.wikipedia.org/wiki/Albert_Einstein",'6.2.1 Albert.txt')
```
